## COUNT or FREQUENCY ENCODING


### Due to high cardinality(many number of labels)

- To replace each label of the categorical variable by the count ,this is the amount of times each labels appears in the dataset




```python
import pandas as pd
import numpy as np
df=pd.read_csv("train_m.csv",usecols=['X1','X2'])
```

```python
df.head()
```

```python
df.shape
```

```python
# one hot encoding
pd.get_dummies(df).shape
```

```python
len(df['X1'].unique())
```

```python
for i in df.columns[0:]:
    print('unique values of',i,':',len(df[i].unique()))
```

```python
df_dict=df.X2.value_counts().to_dict()
```

```python
#mapping the categorical counts
df.X2=df.X2.map(df_dict)
```

##  How to handle ordinal categorical variables



```python
import pandas as pd
import datetime
```

```python
dfbase=datetime.datetime.today()
dfdatelist=[dfbase-datetime.timedelta(days=x)for x in range (0,20)]
```

```python
df=pd.DataFrame(dfdatelist)
df.columns=['Day']
```

```python
df
```

```python
df['dayoftheweek']=df['Day'].dt.day_name()
```

```python
df.head()
```

```python
weekday_map = {'Monday':1,
               'Tuesday':2,
               'Wednesday':3,
               'Thursday':4,
               'Friday':5,
               'Saturday':6,
               'Sunday':7
}
df['day_ordinal'] = df.dayoftheweek.map(weekday_map)
df.head(20)
df.drop('dayof the week',axis=1)
```
